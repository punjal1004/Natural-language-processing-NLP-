```python
import pandas as pd
data = pd.read_csv('/content/gdrive/My Drive/dataset/Dataset.csv' , sep = '\t' , names = ['label'], header = None)
data.head()
```

**Pre-Processing Data**

**Remove punctuation**

```python
import string
string.punctuation
```

```python
def remove_punct(text):
  text_nopunct = "".join([char for char in text if char not in string.punctuation])
  return text_nopunct
data['label_clean'] = data['label'].apply(lambda x: remove_punct(x))
data.head()
```

**Tokenization**

```python
import re
def tokenize(text):
  tokens = re.split('\W+' , text)
  return tokens
data['label_tokenized'] = data['label_clean'].apply(lambda x: tokenize(x))
data.head()
```

**Remove stopwords**

```python
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
  text = [word for word in tokenized_list if word not in stopword]
  return text
data['label_nostop'] = data['label_tokenized'].apply(lambda x: remove_stopwords(x))
data.head()
```

**Stemming**

```python
ps = nltk.PorterStemmer()
def stemming(tokenized_text):
  text = [ps.stem(word) for word in tokenized_text]
  return text
data['label_stemmed'] = data['label_nostop'].apply(lambda x: stemming(x))
data.head()
```

**Lemmatizing**

```python
wn =  nltk.WordNetLemmatizer()
def lemmatizing(tokenized_text):
  text = [wn.lemmatize(word) for word in tokenized_text]
  return text
data['label_lemmetized'] = data['label_nostop'].apply(lambda x: stemming(x))
data.head()
```
